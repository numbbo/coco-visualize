```python
import cocoviz as vz
import cocopp as pp

from cocoviz import ResultSet

import polars as pl
```

```python
def read_coco_dataset(name):
    resultset = ResultSet()
    data = pp.load2(name)
    for result in data:
        evals = result.evals
        for i in range(1, evals.shape[1]):
            problem = vz.ProblemDescription(f"bbob-f_{result.funcId}", result.instancenumbers[i - 1], result.dim, 1)            
            r = vz.Result(result.algId, problem, pl.DataFrame({"value": evals[:, 0], "fevals": evals[:, i]}).drop_nans())            
            resultset.append(r)
    return resultset
```

```python
cma = read_coco_dataset("2020/CMA-ES-2019_Hansen.tgz")
powell = read_coco_dataset("2019/Powell-scipy-2019_Varelas.tgz")
bfgs = read_coco_dataset("2019/BFGS-scipy-2019_Varelas.tgz")
nm = read_coco_dataset("2019/Nelder-Mead-scipy-2019_Varelas.tgz")
```

```python
rs = ResultSet().extend(cma).extend(powell).extend(bfgs).extend(nm)

ind =  vz.Indicator("value", larger_is_better=False)
targets = vz.targets.full_targets(rs, ind)
```
